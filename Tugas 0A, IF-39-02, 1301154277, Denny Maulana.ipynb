```python
import pandas as pd

class Hero:
    # menampung satu Hero
    # name, movement speed, physical attack, armor, hp, mana

    def __init__(self, name, movement_speed, physical_attack, armor, hp, mana):
        self.name = name
        self.movement_speed = movement_speed
        self.physical_attack = physical_attack
        self.armor = armor
        self.hp = hp
        self.mana = mana
        self.role = ""

    def inputRole(self):
        # menentukan role dari hero tersebut
        if self.movement_speed >= 270:
            self.role = "MARKSMAN"
        elif self.mana >= 500:
            self.role = "MAGE"
        elif self.armor >= 25:
            self.role = "TANK"
        elif self.physical_attack >= 121:
            self.role = "ASSASSIN"
        elif self.hp >= 2580:
            self.role = "FIGHTER"
        else:
            self.role = "SUPPORT"
            
    def getRole(self):
        return self.role

    def info(self):
        print(self.name, ", ", self.role)

# end class

class Heroes:
    # menampung seluruh hero yang tersedia
    
    hero = []

    def inputHero(self, name, movement_speed, physical_attack, armor, hp, mana):
        h1 = Hero(name, movement_speed, physical_attack, armor, hp, mana)
        h1.inputRole()
        self.hero.append(h1)
        
    def getRole(self, angka):
        return self.hero[angka].getRole()
    
    def info(self):
        print("Name, Role")
        for i in range(0, len(self.hero)):
            self.hero[i].info()

# end class

df = pd.read_csv('D://test.csv')
df
# referensi: https://chrisalbon.com/python/data_wrangling/pandas_dataframe_importing_csv/

name = df.name
movement_speed = df.movementSpeed
physical_attack = df.physicalAttack
armor = df.armor
hp = df.hp
mana = df.mana
role = []

heroes = Heroes()

for i in range(0, len(df)):
    heroes.inputHero(name[i], movement_speed[i], physical_attack[i], armor[i], hp[i], mana[i])
    role.append(heroes.getRole(i))

raw_data = {'name': name,
        'role': role}
df = pd.DataFrame(raw_data, columns = ['name', 'role'])
df
# referensi: https://chrisalbon.com/python/data_wrangling/pandas_dataframe_importing_csv/

df.to_csv('D://prediction.csv', index=False)
# referensi: https://stackoverflow.com/questions/20845213/how-to-avoid-python-pandas-creating-an-index-in-a-saved-csv
```
